Create a Truck Risk Score.

    Formula:

Risk=0.35(overload ratio)+0.25(avg delay)+0.25(fuel inefficiency)+0.15(failure probability)

```python
fuel=distance*fuel_rate[:,None]
overload=np.maximum(weights.sum(axis=1)/capacity-1,0)
delay=delivery_time.mean(axis=1)/100
ineff=fuel.sum(axis=1)/fuel.sum(axis=1).max()
fail=np.random.rand(500)
risk=0.35*overload+0.25*delay+0.25*ineff+0.15*fail
print("Truck Risk Score =")
print(risk)
```

Implement mini NumPy functions manually:

Create your own version of:

    mean
    std
    normalization

Without using NumPy built-ins.

```python
def my_mean(x):
    return x.sum()/x.size

def my_std(x):
    m=my_mean(x)
    return np.sqrt(((x-m)**2).sum()/x.size)

def normalize(x):
    return (x-x.min())/(x.max()-x.min())

print("Custom Mean =",my_mean(distance))
print("Custom Std =",my_std(distance))
print("Normalized Sample =")
print(normalize(distance[:2,:5]))
```
